종목코드(Ticker)와 종목명(Stock Name)을 매칭시킨 데이터 리스트를 csv파일 형태로 저장하기 위한 코드이다.
데이터 분석 시에는 종목코드를 활용해야하고, 시각화할 때는 주로 종목명을 사용하기 때문에
이 두 가지를 매칭시키는 과정은 주가 데이터 분석에 있어서 필수적이다.
우선 정보데이터시스템(KRX) 홈페이지에서 상장기업 상세정보를 검색하여 다운 받은 csv 파일을 해당 파일과 같은 경로에 저장해두었다.
http://data.krx.co.kr/contents/MDC/MAIN/main/index.cmd

```python
import pandas as pd
```

```python
file_path = 'data_5806_20220111.csv'
data = pd.read_csv(file_path, encoding='cp949')
data
```

```python
#필요한 column만 가져오기
data = data[["종목코드", "종목명"]]
data
```

```python
#종목코드를 6자리로 맞춰주기 위한 코드
for i in range(len(data["종목코드"])):
    data["종목코드"].iloc[i] = (6-len(str(data["종목코드"][i])))*"0" + str(data["종목코드"][i])
data
```

```python
data
```

```python
#종목명에 특정 단어가 포함된 리스트만 따로 추출하기. 삼성 계열사를 예시로 들었다.
condition = data["종목명"].str.contains('삼성')
data[condition]
```

```python
data.to_csv("Ticker Match.csv")
```
